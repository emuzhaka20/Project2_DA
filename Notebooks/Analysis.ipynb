```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Loading the dataset
data = pd.read_excel(r'C:\Users\eljo6\Desktop\Project2_DA\Dataset\Dataset.xlsx',)

# Histogram of average scores
plt.figure(figsize=(15, 10))
sns.histplot(data['average_score'], bins=15, kde=True)
plt.title('Distribution of average scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.show()  # Make sure to include this line

# Scatter plot of average scores regarding gender
plt.figure(figsize=(4, 4))
sns.barplot(data=data, x='gender', y='average_score')
plt.title('Gender Effect on average scores')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()

# Box plot of average scores regarding ethnicity, Parental education, lunch and preparation course
plt.figure(figsize=(10, 6))
sns.barplot(x='race_ethnicity', y='average_score', data=data)
plt.title('AVG scores based on race ethnicity')
plt.xlabel('ethnicity')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='parental_level_of_education', y='average_score', data=data)
plt.title('AVG scores based on parental level of education')
plt.xlabel('parental_level_of_education')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(4, 5))
sns.barplot(data=data, x='lunch', y='average_score')
plt.title('Lunch effect on avg score')
plt.xlabel('Lunch')
plt.ylabel('Score')
plt.show()

plt.figure(figsize=(4, 5))
sns.barplot(data=data, x='test_preparation_course', y='average_score')
plt.title('test preparation course effect on avg score')
plt.xlabel('test_preparation_course')
plt.ylabel('Score')
plt.show()
```
