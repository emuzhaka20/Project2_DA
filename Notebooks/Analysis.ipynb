```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots are displayed inline
%matplotlib inline

# Load the dataset (make sure the path is correct)
data = pd.read_excel(r'C:\Users\eljo6\Desktop\Project2_DA\Dataset\Dataset.xlsx',)

# Histogram of Happiness Scores (adjust column name if necessary)
plt.figure(figsize=(15, 10))
sns.histplot(data['average_score'], bins=10, kde=True)
plt.title('Distribution of average scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.show()  # Make sure to include this line

# Scatter plot of Happiness Score vs GDP per Capita
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='gender', y='average_score')
plt.title('Gender Effect on average scores')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()

# Box plot of Happiness Scores by Region (if applicable)
plt.figure(figsize=(25, 6))
sns.barplot(x='race_ethnicity', y='average_score', data=data)
plt.title('AVG scores based on race ethnicity')
plt.xlabel('ethnicity')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='parental_level_of_education', y='math_score', data=data)
plt.title('math scores based on parental_level_of_education')
plt.xlabel('parental_level_of_education')
plt.ylabel('Math Score')
plt.xticks(rotation=45)
plt.show()

# Scatter plot of Happiness Score vs Social Support
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='lunch', y='average_score')
plt.title('Lunch effect on avg score')
plt.xlabel('Lunch')
plt.ylabel('Score')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='test_preparation_course', y='average_score')
plt.title('test_preparation_course effect on avg score')
plt.xlabel('test_preparation_course')
plt.ylabel('Score')
plt.show()
```
